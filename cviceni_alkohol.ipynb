```python
data_men = data_alkohol[data_alkohol['Sex'] == 'Male']
men_counts_and_gpa = data_men.groupby('Faculty').agg(
    male_count=('Sex', 'count'),
    average_gpa=('Yearly_GPA', 'mean')
    ).reset_index()

print("\nPočet mužů na každé fakultě a jejich průměrné Yearly_GPA:")
print(men_counts_and_gpa)
```

Muži mají ve všech zaměřeních fakult podobný průměr známek, nejvíce mužů studuje Ekonomiku a management. Je to opravdu velký počet mužů oproti jiným fakultám. V medicíně a zdravotnictví studují jen 2 muži, což je menší počet než u žen v tomto oboru studiu.

```python
def convert_socialising_frequency(freq_str):
    if freq_str == 'only weekends':
        return 2  # Představuje 2 dny
    elif freq_str == '4+':
        return 5  # Představuje 4 a více
    else:
        try:
            return int(freq_str) 
        except ValueError:
            return None  

data_men = data_alkohol[data_alkohol['Sex'] == 'Male']

data_men['Socialising_Frequency'] = data_men['Socialising_Frequency'].apply(convert_socialising_frequency)

average_socialising_frequency_men = data_men['Socialising_Frequency'].mean()
print("\nPrůměrná frekvence socializace pro muže:")
print(average_socialising_frequency_men)
```

```python
def convert_study_hours(hours_str):
    if hours_str == '1-3':
        return 2  # Průměr mezi 1 a 3
    elif hours_str == '3-5':
        return 4  # Průměr mezi 3 a 5
    elif hours_str == '5-8':
        return 6.5  # Průměr mezi 5 a 8
    elif hours_str == '8+':
        return 8 
    else:
        return None  

data_men = data_alkohol[data_alkohol['Sex'] == 'Male']

data_men['Study_Hours'] = data_men['Study_Hours'].apply(convert_study_hours)

most_common_study_hours = data_men['Study_Hours'].mode()[0]
print(f"\nNejčastější hodnota studijních hodin pro muže: {most_common_study_hours}")
```

I Muži věnují studiu 8 a více hodin za týden. Což je stejné jako u žen

```python
def convert_drinks(drinks_str):
    if drinks_str == '1-3':
        return 2  # Průměr mezi 1 a 3
    elif drinks_str == '3-5':
        return 4  # Průměr mezi 3 a 5
    elif drinks_str == '5-8':
        return 6.5  # Průměr mezi 5 a 8
    elif drinks_str == '8+':
        return 8 
    else:
        return None 

data_women = data_alkohol[data_alkohol['Sex'] == 'Female']
data_men = data_alkohol[data_alkohol['Sex'] == 'Male']

data_women['Drinks_Per_Night'] = data_women['Drinks_Per_Night'].apply(convert_drinks)
data_men['Drinks_Per_Night'] = data_men['Drinks_Per_Night'].apply(convert_drinks)

average_drinks_women = data_women['Drinks_Per_Night'].mean()
average_drinks_men = data_men['Drinks_Per_Night'].mean()

print(f"\nPrůměrný počet drinků za večer pro ženy: {average_drinks_women:.2f}")
print(f"Průměrný počet drinků za večer pro muže: {average_drinks_men:.2f}")
```

Jak se dalo očekávat, ženy za večer vypijí méně drinků než muži.
Muži i ženy chodí do společnosti průměrně 1-2x týdně, spíše 2x.
